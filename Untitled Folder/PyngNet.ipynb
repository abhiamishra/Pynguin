```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True)

train_generator = train_datagen.flow_from_directory(
    'raw_image/train',
    target_size=(64,64),
    batch_size=32,
    class_mode='categorical'
)
```

```python
#First steps in using CNNs
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense
from tensorflow.keras.optimizers import SGD
```

```python
def network():
    cnet = Sequential()
    cnet.add(Conv2D(filters=64, kernel_size=3, activation='relu', input_shape=(64,64,3)))
    cnet.add(MaxPooling2D(pool_size=(2,2)))
    
    cnet.add(Conv2D(filters=64, kernel_size=3, activation='relu', input_shape=(64,64,3)))
    cnet.add(MaxPooling2D(pool_size=(2,2)))
    
    cnet.add(Conv2D(filters=64, kernel_size=3, activation='relu', input_shape=(64,64,3)))
    cnet.add(MaxPooling2D(pool_size=(2,2)))
    
    cnet.add(Flatten())
    
    cnet.add(Dense(64, activation='relu'))
    cnet.add(Dropout(0.25))
    cnet.add(Dense(10, activation='softmax'))
    
    return cnet
```

```python
cnet = network()
opt = SGD(lr=0.01, momentum=0.9)
cnet.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
cnet.fit_generator(
        train_generator,
        steps_per_epoch=50,
        epochs=10)
```
